```python
students = []

while True:
    print("\n1.Add  2.View  3.Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        students.append(name)

    elif choice == "2":
        print("Students:", students)

    elif choice == "3":
        break

    else:
        print("Invalid choice")
```
